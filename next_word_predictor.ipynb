```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
with open('1661-0.txt','r',encoding='utf-8') as file:
  text =file.read()
```

```python
tokenizer=Tokenizer()
tokenizer.fit_on_texts([text])
```

```python
tokenizer.word_index
```

```python
total_words=len(tokenizer.word_index)+1
total_words
```

```python
input_sequences=[]
for sentence in text.split('\n'):
     tokenizer_sentence=tokenizer.texts_to_sequences([sentence])[0]

     for i in range(1,len(tokenizer_sentence)):
      input_sequences.append(tokenizer_sentence[:i+1])


input_sequences
```

```python
max_len=max([len(x) for x in input_sequences])
```

```python
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequence=pad_sequences(input_sequences,maxlen=max_len,padding='pre')
```

```python
X=padded_input_sequence[: ,:-1]
```

```python
y=padded_input_sequence[:,-1]
```

```python
y
```

```python
from tensorflow.keras.utils import to_categorical
y=to_categorical(y,num_classes=total_words)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model = Sequential()
model.add(Embedding(total_words, 100, input_shape=(max_len + 1,)))
model.add(LSTM(150, return_sequences=True))
model.add(LSTM(150))
model.add(Dense(total_words, activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Summary
model.summary()
```

```python
model.fit(X,y,epochs=50)
```
